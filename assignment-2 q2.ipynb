```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
a = np.array([65,82,81,67,57,59,66,75,82,70])
popmean = a.mean()
popstd = a.std()
mean = []
std = []
for i in range(20):
    b = np.random.default_rng(i)
    c = b.choice(a,size =3,replace = False)
    mean.append(c.mean())
    std.append(c.std())
iterations  = np.arange(20)+1
popmean1 = np.full(20,popmean)
plt.plot(iterations,mean,label = "Sample Mean")
plt.xticks(iterations)
plt.plot(iterations,popmean1,label ="Population Mean")
plt.xlabel("Number of Iterations" , fontsize =12)
plt.ylabel("Mean Values")
plt.title("Sample mean Vs Population Mean")
plt.legend()
```

```python
popstd1 = np.full(20,popstd)
plt.plot(iterations,std, label ="Sample Standard Deviation")
plt.plot(iterations,popstd1,label="Population Standard Deviation")
plt.xticks(iterations)
plt.xlabel("Number of Iterations" , fontsize =12)
plt.ylabel("STD Values")
plt.title("Sample STD Vs Population STD")
plt.legend()
```

```python
p = [1,  10,  100,  1000 , 10000, 100000]
```

```python
cool = []
for i in p:
    mean = []
    std = []
    a = np.array([65,82,81,67,57,59,66,75,82,70])
    for i in range(i):
        b = np.random.default_rng(i)
        c = b.choice(a,size =3,replace = False)
        mean.append(c.mean())
        std.append(c.std())
    l = np.array(mean)
    cool.append(l.mean())

plt.plot(p,cool,"r")
plt.xlabel("Number of iterations")
plt.ylabel("Cumulative average of whole iteration")
```
